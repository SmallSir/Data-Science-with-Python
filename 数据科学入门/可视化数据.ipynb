### 条形图


```python
from matplotlib import pyplot as plt
```

```python
movies = ['Annie Hall','Ben-Hur','Casablance','Gandhi','West Side Story']
num_oscars = [5, 11, 3, 8 ,10]
```

```python
xs = [i + 0.1 for i in range(len(movies))]
```

```python
plt.bar(xs,num_oscars)
plt.title('我最喜欢的电影')
plt.xlabel('电影名称')
plt.ylabel('所获奥斯卡金像奖数量')
plt.xticks([i + 0.1 for i in range(len(movies))],movies)
plt.axis([2012.5,2014.5,0,10])#范围
plt.show()
```

### 线图

```python
variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_error = [x + y for x,y in zip(variance,bias_squared)]
xs = [i for i in range(len(variance))]
```

```python
plt.plot(xs,variance,'g',label = 'vaniance')
plt.plot(xs,total_error,'r',label='sa')
plt.legend(loc = 9)
plt.show()
```

### 散点图

```python
friends = [70,65,72,63,71,64,60,64,67]
minutes = [175,170,205,120,220,130,105,145,190]
labels = ['a','b','c','d','e','f','g','h','i']
plt.scatter(friends,minutes)
#给每个点加标记
for label,friend_count,minute_count in zip(labels,friends,minutes):
    plt.annotate(label,xy=(friend_count,minute_count),#把标记放在对应的点上
                xytext = (5,-5),#但要有轻微偏离
                textcoords = 'offset points'
                )
plt.show()
```
